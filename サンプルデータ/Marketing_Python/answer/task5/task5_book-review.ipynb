## STEP3: 抽出結果の整形

```python
words = []
for token in t.tokenize(review_sample):
    if (token.part_of_speech.split(',')[0] == '名詞') & (token.part_of_speech.split(',')[1] in ['一般', '固有名詞', 'サ変接続', '形容動詞語幹']):
        words.append(token.surface)
words
```

```python
words_list = []
for review in book1_kuchikomi_df['レビュー']:
    words = []
    for token in t.tokenize(review):
        if (token.part_of_speech.split(',')[0] == '名詞') & (token.part_of_speech.split(',')[1] in ['一般', '固有名詞', 'サ変接続', '形容動詞語幹']):
            words.append(token.surface)
    words_list.append(words)
words_list
```

```python
def get_words_list(_kuchikomi_df):
    words_list = []
    for review in _kuchikomi_df['レビュー']:
        words = []
        for token in t.tokenize(review):
            if (token.part_of_speech.split(',')[0] == '名詞') & (token.part_of_speech.split(',')[1] in ['一般', '固有名詞', 'サ変接続', '形容動詞語幹']):
                words.append(token.surface)
        words_list.append(words)
    return words_list
```

```python
book1_words_list = get_words_list(book1_kuchikomi_df)
book1_words_list
```

```python
book2_words_list = get_words_list(book2_kuchikomi_df)
book2_words_list
```

```python
book1_kuchikomi_df['名詞'] = book1_words_list
book1_kuchikomi_df
```

```python
book2_kuchikomi_df['名詞'] = book2_words_list
book2_kuchikomi_df
```

# 抽出した単語から口コミの傾向を調べる

## STEP1: 商品別の名詞の出現頻度分析

```python
import collections
```

```python
counter = collections.Counter(['犬', '猫', '犬', '猫', '鳥', '猫'])
counter
```

```python
counter.most_common()
```

```python
all_words_book1 = []
for words in book1_kuchikomi_df['名詞']:
    all_words_book1.extend(words)
```

```python
len(all_words_book1)
```

```python
book1_counter = collections.Counter(all_words_book1)
book1_counter.most_common()
```

```python
all_words_book2 = []
for words in book2_kuchikomi_df['名詞']:
    all_words_book2.extend(words)

book2_counter = collections.Counter(all_words_book2)
book2_counter.most_common()
```

## STEP2: 評価別の名詞の出現頻度分析

```python
all_words_good = []
for words in book1_kuchikomi_df[book1_kuchikomi_df['評価']=='満足']['名詞']:
    all_words_good.extend(words)
for words in book2_kuchikomi_df[book2_kuchikomi_df['評価']=='満足']['名詞']:
    all_words_good.extend(words)

good_counter = collections.Counter(all_words_good)
good_counter.most_common()
```

```python
all_words_notgood = []
for words in book1_kuchikomi_df[book1_kuchikomi_df['評価']=='不満足']['名詞']:
    all_words_notgood.extend(words)
for words in book2_kuchikomi_df[book2_kuchikomi_df['評価']=='不満足']['名詞']:
    all_words_notgood.extend(words)

notgood_counter = collections.Counter(all_words_notgood)
notgood_counter.most_common()
```
