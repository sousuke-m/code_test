```python
# ライブラリインポート
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# ページ情報の取得
url='https://www.imdb.com/search/keyword/?keywords=public-domain&ref_=kw_ref_typ&sort=moviemeter,asc&mode=detail&page=1&title_type=movie'
response =requests.get(url)
response.status_code
```

```python
# BeautifulSoupの読み込み
soup = BeautifulSoup(response.text, 'html.parser')

# 映画リストの抜き出し
movies = soup.find_all('div', {'class': 'lister-item mode-detail'})
```

```python
# 自作関数に例外処理を追加
def movie_scraping(movie):
    # title の取得
    title = movie.find('h3', {'class': 'lister-item-header'}).find('a').text
    # yearの取得
    raw_year = movie.find('h3', {'class': 'lister-item-header'}).find('span', {'class': 'lister-item-year text-muted unbold'}).text
    year = raw_year.lstrip('(').rstrip(')')
    # runtimeの取得
    try:
        runtime = movie.find('p', {'class': 'text-muted text-small'}).find('span', {'class': 'runtime'}).text
    except:
        runtime = ''
    # genreの取得
    try:
        raw_genre = movie.find('p', {'class': 'text-muted text-small'}).find('span', {'class': 'genre'}).text
        genre = raw_genre.strip()
    except:
        genre = ''
    # directorとcastの取得
    director = ''
    cast = ''
    count = 0
    for i in movie.find_all('p', {'class': 'text-muted text-small'})[1].find_all('a'):
        if count == 0:
            director = i.text
        else:
            cast += i.text + ','
        count += 1
    # gross の取得
    try:
        gross = movie.find_all('p', {'class': 'text-muted text-small'})[2].find_all('span', {'name': 'nv'})[1]['data-value']
    except:
        gross = ''
    
    return {'title': title, 'year': year, 'runtime': runtime, 'genre': genre, 'director': director, 'cast': cast, 'gross': gross}
```

---
##### 解説

*  `year = raw_year.lstrip('(').rstrip(')')` の処理について

raw_year をそのまま辞書にいれてCSVに出力してみてください。<br>
もしかすると、ExcelでCSVを開いた際に、年がマイナスで表示されているかもしれません。<br>
もしマイナス表記で表示されてしまう場合、 `(数値)` を自動でマイナス表記に変換するというExcel上の仕様になります。<br>
この場合、Excelのセルの書式設定で `(数値)` をそのまま表示する設定にするか、 `()` を取ってしまうのが手っ取り早いです。<br>
文頭にある `(` は、 `.lstrip('(')` と指定することで削除することができます。<br>
同様に文末にある `)` は `.rstrip(')')` で削除することができます。<br>
これらはPythonにデフォルトで実装されている関数になっています。ご自身で調べて実装することはできたでしょうか？

*  `genre = raw_genre.strip()` の処理について

raw_genreをそのまま辞書に入れてみてください。<br>
下記でdfを表示した際に、 先頭に `\n` があったり、文末に余分は半角スペースが含まれており、我慢ならない人もいるのではないでしょうか？笑<br>
そんな几帳面な読者の方は、 `.strip()` とすることで、　文頭・文末から余計な `\n` や 半角スペースを削除することができます。<br>
この関数も、Pythonにデフォルトで実装されている関数になっています。

* cast の取得処理について

castを上手に取得することはできましたか？本節までの学びを活かせば必ず取得できるはずです。<br>
本書のP239-240でdirectorを取得した際、出演者（stars）まで取得されていました。<br>
find_all関数を使用するとリスト形で取得できるんでしたね？<br>
リストで取得したものに処理を加えたい場合は、for文を使って処理をするのが基本になります。<br>
本書で説明した様に、リストの0番目にdirectorが入るため、if文で制御して1番目以降がcast変数に追加される様に処理を追加した形になっています。<br>

```python
records = []
for movie in movies:
    temporary_data = movie_scraping(movie)
    records.append(temporary_data)

# 表示して確認
records
```
